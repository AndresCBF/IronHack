### Clean DataSet deaths related to alcohol

Treat NaNs in death columns and convert all numerical values in positive integers values

```python
def num(value):
    """
    This funtion return any numeircal value as a positive integer number and any NaNs values as integer 0
    
    Example :
        Input: "-9.539797"
        Return: "10"
        
    Example1 :
        Input: "NaN"
        Return: "0"
    """
    
    if value < 0 :
        value = -value
        return int(round(value,0))
    elif value >= 0 :
        return int(round(value,0))
    else :
        value = 0
        return int(round(value,0))
```

```python
alcoholdeath['Deaths_Under_5']= alcoholdeath['Deaths_Under_5'].apply(num)
alcoholdeath['Deaths_5-14_years']= alcoholdeath['Deaths_5-14_years'].apply(num)
alcoholdeath['Deaths_15-49_years']= alcoholdeath['Deaths_15-49_years'].apply(num)
alcoholdeath['Deaths_50-69_years']= alcoholdeath['Deaths_50-69_years'].apply(num)
alcoholdeath['Deaths_70+_years']= alcoholdeath['Deaths_70+_years'].apply(num)
# Create Colum Total Deaths 
alcoholdeath['Total_deaths'] = alcoholdeath['Deaths_Under_5']+alcoholdeath['Deaths_5-14_years']+alcoholdeath['Deaths_15-49_years']+alcoholdeath['Deaths_50-69_years']+alcoholdeath['Deaths_70+_years']
alcoholdeath
```

```python
alcoholdeathperyear = alcoholdeath.groupby(['Year']).agg(
    total_global_deaths = pd.NamedAgg('Total_deaths', 'sum'))
alcoholdeathperyear = alcoholdeath.merge(alcoholdeathperyear, on=['Year'])
alcoholdeathperyear
```

Fullfill NaNs in column "Country_cod"  

```python
paises = alcoholdeath[['Country', 'Country_cod']].dropna()
paises = paises.groupby('Country').first()
alcoholdeath = pd.merge(alcoholdeath, paises, on = ['Country', 'Country_cod'])
alcoholdeathper=alcoholdeath.drop(alcoholdeath[alcoholdeath['Country']=='World'])
#alcoholdeathper['per_total_deaths']=round(alcoholdeathper['Total_deaths']/alcoholdeathper['total_global_deaths'],6)
alcoholdeathper
```
