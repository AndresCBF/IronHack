## DataSet global deaths related to alcohol 

### Review and read datset  

```python
file = open('C:/Users/forqu/OneDrive/Documentos/IronHack/Github/Projetos/History of Rock(1950-2020)/Data/deaths-attributed-to-alcohol-use-by-age.csv', 'rb')
file_lines = file.readlines(10000)
for i in range(5):
    print(file_lines[i].decode('latin-1'))
file.close()
alcoholdeath= pd.read_csv('C:/Users/forqu/OneDrive/Documentos/IronHack/Github/Projetos/History of Rock(1950-2020)/Data/deaths-attributed-to-alcohol-use-by-age.csv', sep = ',', decimal = '.', encoding = 'latin-1')
```

### Clean DataSet deaths related to alcohol

```python
alcoholdeath = alcoholdeath.rename(columns={"Entity":"Country","Code":"Country_cod","Deaths - Alcohol use - Sex: Both - Age: Under 5 (Number)":"Deaths_Under_5", "Deaths - Alcohol use - Sex: Both - Age: 70+ years (Number)": "Deaths_70+_years", "Deaths - Alcohol use - Sex: Both - Age: 50-69 years (Number)":"Deaths_50-69_years","Deaths - Alcohol use - Sex: Both - Age: 5-14 years (Number)":"Deaths_5-14_years","Deaths - Alcohol use - Sex: Both - Age: 15-49 years (Number)":"Deaths_15-49_years"})
alcoholdeath = alcoholdeath[['Country_cod','Country','Year','Deaths_Under_5','Deaths_5-14_years','Deaths_15-49_years','Deaths_50-69_years','Deaths_70+_years']]
alcoholdeath
```

Treat NaNs in death columns and convert all numerical values in positive integers values

```python
def num(value):
    """
    This funtion return any numeircal value as a positive integer number and any NaNs values as integer 0
    
    Example :
        Input: "-9.539797"
        Return: "10"
        
    Example1 :
        Input: "NaN"
        Return: "0"
    """
    
    if value < 0 :
        value = -value
        return int(round(value,0))
    elif value >= 0 :
        return int(round(value,0))
    else :
        value = 0
        return int(round(value,0))
```

```python
alcoholdeath['Deaths_Under_5']= alcoholdeath['Deaths_Under_5'].apply(num)
alcoholdeath['Deaths_5-14_years']= alcoholdeath['Deaths_5-14_years'].apply(num)
alcoholdeath['Deaths_15-49_years']= alcoholdeath['Deaths_15-49_years'].apply(num)
alcoholdeath['Deaths_50-69_years']= alcoholdeath['Deaths_50-69_years'].apply(num)
alcoholdeath['Deaths_70+_years']= alcoholdeath['Deaths_70+_years'].apply(num)
# Create Colum Total Deaths 
alcoholdeath['Total_deaths'] = alcoholdeath['Deaths_Under_5']+alcoholdeath['Deaths_5-14_years']+alcoholdeath['Deaths_15-49_years']+alcoholdeath['Deaths_50-69_years']+alcoholdeath['Deaths_70+_years']
alcoholdeath
```

Fullfill NaNs in column "Country_cod"  

```python
paises = alcoholdeath[['Country', 'Country_cod']].dropna()
paises = paises.groupby('Country').first()
alcoholdeath = pd.merge(alcoholdeath, paises, on = ['Country', 'Country_cod'])
alcoholdeath
```
