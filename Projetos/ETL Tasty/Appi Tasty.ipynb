## Request all recipes per tag, and  create a DataFrame with all recipes 

```python
recipe_id=[]
recipe_name=[]
recipe_score=[]
recipe_positivevotes=[]
recipe_negativevotes=[]
recipe_totalvotes=[]
recipe_country=[]
recipe_servings=[]
recipe_instructions=[]
recipe_totalingredients=[]
ingredient_text=[]
ingredient_id=[]
ingredient_name=[]
ingredient_quantity=[]
unit_id=[]
unit_name=[]
Recipe_calories=[]
Recipe_carbohydrates=[]
Recipe_protein=[]
Recipe_fat=[]
Recipe_fiber=[]
Recipe_sugar=[]
Recipe_video=[]
Recipe_tags=[]
for tag in tags_name :
    url = "https://tasty.p.rapidapi.com/recipes/list"
    querystring = {"from":"0","size":"100","tags":tag}
    response = requests.request("GET", url, headers=headers, params=querystring)
    recipes=json.loads(response.text)
    if len(recipes['results']) > 0 :
        for i in range(len(recipes['results'])) :
            if 'recipes' not in recipes['results'][i].keys() :
                recipeid = recipes['results'][i]['id']
                recipename = recipes['results'][i]['name']
                recipescore = recipes['results'][i]['user_ratings']['score']
                recipepositivevotes = recipes['results'][i]['user_ratings']['count_positive']
                recipenegativevotes = recipes['results'][i]['user_ratings']['count_negative']
                recipetotalvotes = recipes['results'][i]['user_ratings']['count_positive'] + recipes['results'][i]['user_ratings']['count_negative']
                recipecountry = recipes['results'][i]['country']
                recipeservings = recipes['results'][i]['num_servings']
                recipevideo = recipes['results'][i]['original_video_url']
                recipetag=[tag['name'] for tag in recipes['results'][i]['tags']]
                recipeinstructions= ' '.join([instruction['display_text'] for instruction in recipes['results'][i]['instructions']])
                recipetotalingredients = len(recipes['results'][i]['sections'][0]['components'])
                for j in range(len(recipes['results'][i]['sections'][0]['components'])) :
                    ingredienttext = recipes['results'][i]['sections'][0]['components'][j]['raw_text']
                    ingredientid = recipes['results'][i]['sections'][0]['components'][j]['ingredient']['id']
                    ingredientname = recipes['results'][i]['sections'][0]['components'][j]['ingredient']['name']
                    try :                   
                        ingredientquantity = recipes['results'][i]['sections'][0]['components'][j]['measurements'][0]['quantity']
                        unitid = recipes['results'][i]['sections'][0]['components'][j]['measurements'][0]['id']
                        unitname = recipes['results'][i]['sections'][0]['components'][j]['measurements'][0]['unit']['name']
                    except IndexError :
                        ingredientquantity = None
                        unitid = None
                        unitname = None
                    if bool(recipes['results'][i]['nutrition']):
                        calories = recipes['results'][i]['nutrition']['calories'] 
                        carbohydrates = recipes['results'][i]['nutrition']['carbohydrates'] 
                        protein = recipes['results'][i]['nutrition']['protein'] 
                        fat = recipes['results'][i]['nutrition']['fat'] 
                        fiber = recipes['results'][i]['nutrition']['fiber'] 
                        sugar = recipes['results'][i]['nutrition']['sugar']
                    else :
                        calories = None 
                        carbohydrates = None 
                        protein = None 
                        fat = None 
                        fiber = None 
                        sugar = None
                    ingredient_text.append(ingredienttext)
                    ingredient_id.append(ingredientid)
                    ingredient_name.append(ingredientname)
                    ingredient_quantity.append(ingredientquantity)
                    unit_id.append(unitid)
                    unit_name.append(unitname)
                    Recipe_calories.append(calories)
                    Recipe_carbohydrates.append(carbohydrates)
                    Recipe_protein.append(protein)
                    Recipe_fat.append(fat)
                    Recipe_fiber.append(fiber)
                    Recipe_sugar.append(sugar)
                    recipe_id.append(recipeid)
                    recipe_name.append(recipename)
                    recipe_score.append(recipescore)
                    recipe_positivevotes.append(recipepositivevotes)
                    recipe_negativevotes.append(recipenegativevotes)
                    recipe_totalvotes.append(recipetotalvotes)
                    recipe_country.append(recipecountry)
                    recipe_servings.append(recipeservings)
                    recipe_instructions.append(recipeinstructions)
                    recipe_totalingredients.append(recipetotalingredients)
                    Recipe_video.append(recipevideo)
                    Recipe_tags.append(recipetag)
            else :
                for k in range(len(recipes['results'][i]['recipes'])) :
                    recipeid = recipes['results'][i]['recipes'][k]['id']
                    recipename = recipes['results'][i]['recipes'][k]['name']
                    recipescore = recipes['results'][i]['recipes'][k]['user_ratings']['score']
                    recipepositivevotes = recipes['results'][i]['recipes'][k]['user_ratings']['count_positive']
                    recipenegativevotes = recipes['results'][i]['recipes'][k]['user_ratings']['count_negative']
                    recipetotalvotes = recipes['results'][i]['recipes'][k]['user_ratings']['count_positive'] + recipes['results'][i]['recipes'][k]['user_ratings']['count_negative']
                    recipecountry = recipes['results'][i]['recipes'][k]['country']
                    recipeservings = recipes['results'][i]['recipes'][k]['num_servings']
                    recipevideo = recipes['results'][i]['recipes'][k]['original_video_url']
                    recipetag=[tag['name'] for tag in recipes['results'][i]['recipes'][k]['tags']]
                    recipeinstructions= ' '.join([instruction['display_text'] for instruction in recipes['results'][i]['recipes'][k]['instructions']])
                    recipetotalingredients = int(len(recipes['results'][i]['recipes'][k]['sections'][0]['components']))
                    for j in range(len(recipes['results'][i]['recipes'][k]['sections'][0]['components'])) :
                        ingredienttext = recipes['results'][i]['recipes'][k]['sections'][0]['components'][j]['raw_text']
                        ingredientid = recipes['results'][i]['recipes'][k]['sections'][0]['components'][j]['ingredient']['id']
                        ingredientname = recipes['results'][i]['recipes'][k]['sections'][0]['components'][j]['ingredient']['name']
                        try :
                            ingredientquantity = recipes['results'][i]['recipes'][k]['sections'][0]['components'][j]['measurements'][0]['quantity']
                            unitid = recipes['results'][i]['recipes'][k]['sections'][0]['components'][j]['measurements'][0]['id']
                            unitname = recipes['results'][i]['recipes'][k]['sections'][0]['components'][j]['measurements'][0]['unit']['name']
                        except IndexError :
                            ingredientquantity = None
                            unitid = None
                            unitname = None 
                        if bool(recipes['results'][i]['recipes'][k]['nutrition']) :
                            calories = recipes['results'][i]['recipes'][k]['nutrition']['calories'] 
                            carbohydrates = recipes['results'][i]['recipes'][k]['nutrition']['carbohydrates'] 
                            protein = recipes['results'][i]['recipes'][k]['nutrition']['protein'] 
                            fat = recipes['results'][i]['recipes'][k]['nutrition']['fat'] 
                            fiber = recipes['results'][i]['recipes'][k]['nutrition']['fiber'] 
                            sugar = recipes['results'][i]['recipes'][k]['nutrition']['sugar']
                        else :
                            calories = None 
                            carbohydrates = None 
                            protein = None 
                            fat = None 
                            fiber = None 
                            sugar = None
                        ingredient_text.append(ingredienttext)
                        ingredient_id.append(ingredientid)
                        ingredient_name.append(ingredientname)
                        ingredient_quantity.append(ingredientquantity)
                        unit_id.append(unitid)
                        unit_name.append(unitname)
                        Recipe_calories.append(calories)
                        Recipe_carbohydrates.append(carbohydrates)
                        Recipe_protein.append(protein)
                        Recipe_fat.append(fat)
                        Recipe_fiber.append(fiber)
                        Recipe_sugar.append(sugar)
                        recipe_id.append(recipeid)
                        recipe_name.append(recipename)
                        recipe_score.append(recipescore)
                        recipe_positivevotes.append(recipepositivevotes)
                        recipe_negativevotes.append(recipenegativevotes)
                        recipe_totalvotes.append(recipetotalvotes)
                        recipe_country.append(recipecountry)
                        recipe_servings.append(recipeservings)
                        recipe_instructions.append(recipeinstructions)
                        recipe_totalingredients.append(recipetotalingredients)
                        Recipe_video.append(recipevideo)
                        Recipe_tags.append(recipetag)
        df_recepies = pd.DataFrame({'Recipe_id':recipe_id,'Recipe_name':recipe_name,'Recipe_score':recipe_score,'Recipe_positivevotes':recipe_positivevotes,'Recipe_negativevotes':recipe_negativevotes,
                                    'Recipe_totalvotes':recipe_totalvotes,'Recipe_country':recipe_country,'Recipe_servings':recipe_servings,'Recipe_instructions':recipe_instructions,'Recipe_totalingredients':recipe_totalingredients,
                                    'Ingredient_text':ingredient_text,'Ingredient_id':ingredient_id,'Ingredient_name':ingredient_name,'Ingredient_quantity':ingredient_quantity,'Unit_id':unit_id,
                                    'Unit_name':unit_name,'Recipe_calories':Recipe_calories,'Recipe_carbohydrates':Recipe_carbohydrates,'Recipe_protein':Recipe_protein,'Recipe_fat':Recipe_fat,
                                    'Recipe_fiber':Recipe_fiber,'Recipe_sugar':Recipe_sugar,'Recipe_video':Recipe_video,'Recipe_tags':Recipe_tags})
```

```python
def  country (tags):
    '''
    Function return a country name, form a list of tags in a recipe 
    
    Example1:
    tags = ['spatula', 'mixing_bowl', 'sauce_pan', 'desserts', 'deep_fry', 'indulgent_sweets', 
            'stove_top', 'sieve', 'tongs', 'dry_measuring_cups', 'liquid_measuring_cup', 
            'american', 'measuring_spoons', 'snacks', 'easy', 'kid_friendly', 'comfort_food', 
            'chefs_knife', 'cutting_board', 'pyrex', '3_musketeers']
    return = United States
    Example2:
    tag = ['chinese', 'stove_top', 'chefs_knife', 'indulgent_sweets', 'measuring_spoons', 
           'saute_pan', 'liquid_measuring_cup', 'plastic_wrap', 'spatula', 'mixing_bowl', 
           'lunar_new_year', 'desserts']
    return = China
    '''
    tag_country={'vietnamese':'Vietnam','venezuelan':'Venezuela','thai':'Thailand','taiwanese':'Taiwan',
                 'swedish':'Sweden','south_african':'South Africa','puerto_rican':'Puerto Rico','peruvian':'Peru',
                 'persian':'Iran','mexican':'Mexico','lebanese':'Lebanon', 'latin_american':'Latin America',
                 'laotian':'Laos','korean':'Korea','kenyan':'Kenya','japanese':'Japan','jamaican':'Jamaica',
                 'italian':'Italy','indian':'India','hawaiian':'United States','haitian':'Haiti', 'greek':'Greece',
                 'german':'Germany','french':'France','filipino':'Philippines','ethiopian':'Ethiopia',
                 'dominican':'Dominican Republic','cuban':'Cuba','chinese':'China','caribbean':'Caribbean Islands',
                 'british':'United Kingdom','brazilian':'Brazil', 'american':'United States', 'african':'Africa'}
    
    for tag in eval(tags):
        if tag in tag_country.keys():
            country = tag_country[tag]
            return country
```

```python
df_recepies['Country_tags'] = df_recepies['Recipe_tags'].apply(country)
```
