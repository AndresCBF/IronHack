## Subplots
- Note that subplot and subplots are different functions. You can check [here](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.subplots.html).

### Use `plt.subplots(nrows=1, ncols=2)` to plot (x, y) and (x, z)

```python
# Your code here
fig,ax = plt.subplots(nrows=1, ncols=2) 
ax[0].plot(x,y)
ax[1].plot(x,z)
```

### Resize your plot
- Using the plot above add the `figsize` argument in `plt.subplots()` and resize your plot.
- Try to add a `title` and `label` the axis

```python
# Your code here
fig,ax = plt.subplots(nrows=1, ncols=2, figsize=(20,5)) 
ax[0].plot(x,y, label='lienar')
ax[0].set_title('Linear')
ax[0].set_xlabel('numbers')
ax[0].set_ylabel('numbers by 2')
ax[1].plot(x,z, label='quadratic')
ax[1].set_title('Quadratic')
ax[1].set_xlabel('numbers')
ax[1].set_ylabel('numbers Square')
```

## Plot (x, y) and (x, z) in the same axes
- Try to add a legend to the plot

```python
# Your code here
plt.plot(x,y, label='Linear')
plt.plot(x,z, label='Quadratic')
plt.xlabel('numbers')
plt.ylabel('Y label')
plt.title('Plots Linear and  Quadratic')
plt.legend()
plt.show()
```
